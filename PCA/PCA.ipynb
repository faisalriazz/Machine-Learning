```python
import pandas as pd
import numpy as np
import os
df=pd.read_csv('/home/ishaan/Machine-Learning/datasets/sonar.csv')
df=df.reindex(df.index.delete(0))
del df['Class']
print df.head(10)
```

```python
df=df.apply(lambda x: pd.to_numeric(x, errors='ignore'))
```

```python
#mean normalization and feature scaling
df_norm = (df - df.mean()) / (df.max() - df.min())
cov_matrix=df_norm.cov()
```

```python
print cov_matrix
```

```python
#doing single value decomposition to find eigen vectors corresponding to all the eigen values
U, sigma, V = np.linalg.svd(df_norm)
```
